```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np

from wordcloud import WordCloud
```

```python
obam = open('data/obama.csv').read()
tony = open('data/tony.csv').read()
nix = open('data/nixon.csv').read()
roos = open('data/roose.csv').read()
trump = open('data/trump.csv').read()
```

```python
w_obam = WordCloud().generate(obam)
plt.figure(figsize=(20,12))
plt.imshow(w_obam, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
w_nixon = WordCloud().generate(nix)
plt.figure(figsize=(20,12))
plt.imshow(w_nixon, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
w_trump = WordCloud().generate(trump)
plt.figure(figsize=(20,12))
plt.imshow(w_trump, interpolation='bilinear')
plt.axis('off')
plt.show()
```
